## 2. Exploración de datos (EDA)

```python
def list_images(folder):
    exts = {'.jpg', '.jpeg', '.png', '.bmp'}
    rows = []
    for cls_dir in sorted(folder.iterdir()):
        if cls_dir.is_dir():
            for p in cls_dir.iterdir():
                if p.suffix.lower() in exts:
                    rows.append({'path': str(p), 'class': cls_dir.name})
    return pd.DataFrame(rows)

df_train_full = list_images(TRAIN_DIR)
df_test = list_images(TEST_DIR)
print('Total train+val:', len(df_train_full), '| test:', len(df_test))
print('\nDistribución train+val:'); print(df_train_full['class'].value_counts())
print('\nDistribución test:');      print(df_test['class'].value_counts())

# Normalizamos nombres de clase a 'cracked' / 'normal'
MAP = {c: ('cracked' if 'crack' in c.lower() or 'defect' in c.lower() else 'normal')
       for c in df_train_full['class'].unique()}
print('\nMapeo de carpetas:', MAP)
df_train_full['label'] = df_train_full['class'].map(MAP).map({'normal': 0, 'cracked': 1})
df_test['label']       = df_test['class'].map(MAP).map({'normal': 0, 'cracked': 1})

fig, axes = plt.subplots(2, 4, figsize=(14, 6))
for i, lbl_name in enumerate(['normal', 'cracked']):
    samples = df_train_full[df_train_full['class'].map(MAP) == lbl_name].sample(4, random_state=SEED)
    for j, (_, row) in enumerate(samples.iterrows()):
        img = Image.open(row['path']).convert('RGB')
        axes[i, j].imshow(img); axes[i, j].axis('off')
        axes[i, j].set_title(f"{lbl_name}")
plt.suptitle('Muestras del dataset', fontsize=14); plt.tight_layout(); plt.show()
```

## 3. Partición estratificada train/val + Dataset y DataLoaders

```python
df_tr, df_val = train_test_split(
    df_train_full, test_size=0.15, stratify=df_train_full["label"], random_state=SEED)
print("train:", len(df_tr), "val:", len(df_val), "test:", len(df_test))

IMG_SIZE = 224
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

train_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE + 16, IMG_SIZE + 16)),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.85, 1.0)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.RandomRotation(degrees=15),
    transforms.ColorJitter(brightness=0.15, contrast=0.15),
    transforms.RandomGrayscale(p=0.1),
    transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 2.0)),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])
eval_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

class TireDataset(Dataset):
    def __init__(self, df, transform):
        self.df = df.reset_index(drop=True); self.transform = transform
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = Image.open(row["path"]).convert("RGB")
        return self.transform(img), torch.tensor(row["label"], dtype=torch.float32), row["path"]

BATCH = 32
loaders = {
    "train": DataLoader(TireDataset(df_tr,   train_tf), batch_size=BATCH, shuffle=True,  num_workers=NUM_WORKERS),
    "val":   DataLoader(TireDataset(df_val,  eval_tf),  batch_size=BATCH, shuffle=False, num_workers=NUM_WORKERS),
    "test":  DataLoader(TireDataset(df_test, eval_tf),  batch_size=BATCH, shuffle=False, num_workers=NUM_WORKERS),
}
for k, dl in loaders.items():
    print(k, "->", len(dl.dataset), "imgs,", len(dl), "batches")
```

## 4. Arquitectura 1 — CNN entrenada desde cero (`TireCNN`)

Cuatro bloques `Conv-BN-ReLU` × 2 + `MaxPool`, *global average pooling* y cabeza FC con dropout.

```python
class ConvBlock(nn.Module):
    def __init__(self, in_c, out_c):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_c, out_c, 3, padding=1, bias=False), nn.BatchNorm2d(out_c), nn.ReLU(inplace=True),
            nn.Conv2d(out_c, out_c, 3, padding=1, bias=False), nn.BatchNorm2d(out_c), nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
        )
    def forward(self, x): return self.block(x)

class TireCNN(nn.Module):
    def __init__(self, dropout=0.4):
        super().__init__()
        self.b1 = ConvBlock(3, 32)
        self.b2 = ConvBlock(32, 64)
        self.b3 = ConvBlock(64, 128)
        self.b4 = ConvBlock(128, 256)
        self.gap = nn.AdaptiveAvgPool2d(1)
        self.head = nn.Sequential(
            nn.Flatten(), nn.Dropout(dropout),
            nn.Linear(256, 128), nn.ReLU(inplace=True), nn.Dropout(dropout),
            nn.Linear(128, 1),
        )
    def forward(self, x):
        x = self.b1(x); x = self.b2(x); x = self.b3(x); x = self.b4(x)
        return self.head(self.gap(x))
    def feature_map(self, x):
        x = self.b1(x); x = self.b2(x); x = self.b3(x); x = self.b4(x)
        return x

m = TireCNN(); n_params = sum(p.numel() for p in m.parameters() if p.requires_grad)
print(f'TireCNN parámetros entrenables: {n_params:,}')
```

## 5. Funciones de pérdida (BCE + Focal) y bucle de entrenamiento

```python
class FocalLoss(nn.Module):
    """Focal loss binaria (Lin et al., 2017)."""
    def __init__(self, gamma=2.0, alpha=0.5):
        super().__init__(); self.gamma = gamma; self.alpha = alpha
    def forward(self, logits, target):
        bce = F.binary_cross_entropy_with_logits(logits, target, reduction='none')
        p = torch.sigmoid(logits)
        pt = p * target + (1 - p) * (1 - target)
        alpha_t = self.alpha * target + (1 - self.alpha) * (1 - target)
        loss = alpha_t * (1 - pt) ** self.gamma * bce
        return loss.mean()

def compute_metrics(y_true, y_prob, threshold=0.5):
    y_pred = (y_prob >= threshold).astype(int)
    return {
        'accuracy':  accuracy_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred, zero_division=0),
        'recall':    recall_score(y_true, y_pred, zero_division=0),
        'f1':        f1_score(y_true, y_pred, zero_division=0),
        'auc':       roc_auc_score(y_true, y_prob) if len(set(y_true)) > 1 else float('nan'),
    }

@torch.no_grad()
def evaluate(model, loader):
    model.eval(); ys, ps, paths = [], [], []
    for x, y, p in loader:
        x = x.to(DEVICE); logits = model(x).squeeze(1)
        prob = torch.sigmoid(logits).cpu().numpy()
        ys.extend(y.numpy()); ps.extend(prob); paths.extend(p)
    ys = np.array(ys); ps = np.array(ps)
    return compute_metrics(ys, ps), ys, ps, paths

def train_model(model, loaders, criterion, epochs=10, lr=1e-3, weight_decay=1e-4,
                params_to_optimize=None, patience=3, label='model'):
    model.to(DEVICE)
    opt = torch.optim.AdamW(params_to_optimize or model.parameters(), lr=lr, weight_decay=weight_decay)
    sched = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=max(epochs, 1))
    use_amp = ON_GPU
    scaler  = torch.amp.GradScaler('cuda', enabled=use_amp)

    history = []; best_f1 = -1.0; best_state = None; bad = 0
    for ep in range(1, epochs + 1):
        model.train(); t0 = time.time(); losses = []
        for x, y, _ in loaders['train']:
            x, y = x.to(DEVICE), y.to(DEVICE)
            opt.zero_grad(set_to_none=True)
            if use_amp:
                with torch.amp.autocast('cuda'):
                    logits = model(x).squeeze(1)
                    loss = criterion(logits, y)
                scaler.scale(loss).backward(); scaler.step(opt); scaler.update()
            else:
                logits = model(x).squeeze(1)
                loss = criterion(logits, y)
                loss.backward(); opt.step()
            losses.append(loss.item())
        sched.step()
        val_metrics, *_ = evaluate(model, loaders['val'])
        history.append({'epoch': ep, 'train_loss': float(np.mean(losses)),
                        **{f'val_{k}': v for k, v in val_metrics.items()}})
        print(f'[{label}] ep {ep:02d} | loss {np.mean(losses):.3f} | '
              f"val F1 {val_metrics['f1']:.3f} | val AUC {val_metrics['auc']:.3f} | {time.time()-t0:.1f}s")
        if val_metrics['f1'] > best_f1:
            best_f1 = val_metrics['f1']
            best_state = {k: v.detach().cpu().clone() for k, v in model.state_dict().items()}
            bad = 0
        else:
            bad += 1
            if bad >= patience:
                print(f'Early stopping en epoch {ep}'); break
    if best_state is not None:
        model.load_state_dict(best_state)
    return model, pd.DataFrame(history)
```

## 6. Entrenamiento del modelo 1 — TireCNN

```python
model_cnn = TireCNN()
criterion = nn.BCEWithLogitsLoss()
model_cnn, hist_cnn = train_model(model_cnn, loaders, criterion,
                                  epochs=EPOCHS_CNN, lr=1e-3, patience=PATIENCE, label='TireCNN')

test_metrics_cnn, y_true_cnn, y_prob_cnn, paths_cnn = evaluate(model_cnn, loaders['test'])
print('\nTireCNN — métricas TEST:', json.dumps(test_metrics_cnn, indent=2))
```

## 7. Modelo 2 — ResNet-50 con fine-tuning gradual

```python
def build_resnet50(freeze_backbone=True):
    net = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
    if freeze_backbone:
        for p in net.parameters(): p.requires_grad = False
    net.fc = nn.Linear(net.fc.in_features, 1)
    return net

# Fase 1: solo cabeza
resnet = build_resnet50(freeze_backbone=True).to(DEVICE)
params_head = [p for p in resnet.parameters() if p.requires_grad]
resnet, hist_r1 = train_model(resnet, loaders, nn.BCEWithLogitsLoss(),
                              epochs=EPOCHS_RESNET_HEAD, lr=1e-3,
                              params_to_optimize=params_head,
                              patience=PATIENCE, label='ResNet50/head')

# Fase 2: descongelar layer3 y layer4, lr más bajo
for name, p in resnet.named_parameters():
    p.requires_grad = name.startswith(('layer3', 'layer4', 'fc'))
params_ft = [p for p in resnet.parameters() if p.requires_grad]
print('Params descongelados para FT:', sum(p.numel() for p in params_ft))

resnet, hist_r2 = train_model(resnet, loaders, nn.BCEWithLogitsLoss(),
                              epochs=EPOCHS_RESNET_FT, lr=1e-4,
                              params_to_optimize=params_ft,
                              patience=PATIENCE, label='ResNet50/FT')

test_metrics_rn, y_true_rn, y_prob_rn, paths_rn = evaluate(resnet, loaders['test'])
print('\nResNet-50 — métricas TEST:', json.dumps(test_metrics_rn, indent=2))

torch.save(model_cnn.state_dict(), CHECKPOINT_DIR / 'tirecnn.pt')
torch.save(resnet.state_dict(),    CHECKPOINT_DIR / 'resnet50.pt')
print('Checkpoints guardados en', CHECKPOINT_DIR)
```

## 9. Estudio de ablación sobre ResNet-50

Comparamos tres factores manteniendo todo lo demás fijo:
- BCE estándar vs Focal Loss (γ=2)
- Con vs sin aumento de datos
- Con vs sin `pos_weight` para mitigar desbalance

```python
def make_resnet_ft_clone():
    """Construye un ResNet-50 listo para FT (cabeza + layer3/4 entrenables)."""
    net = build_resnet50(freeze_backbone=True)
    for name, p in net.named_parameters():
        p.requires_grad = name.startswith(('layer3', 'layer4', 'fc'))
    return net

# Loaders sin aumento (solo eval_tf)
loaders_noaug = dict(loaders)
loaders_noaug['train'] = DataLoader(TireDataset(df_tr, eval_tf), batch_size=BATCH,
                                    shuffle=True, num_workers=NUM_WORKERS)

# pos_weight basado en train
n_neg = int((df_tr['label'] == 0).sum())
n_pos = int((df_tr['label'] == 1).sum())
pos_weight = torch.tensor([n_neg / max(n_pos, 1)], device=DEVICE)
print(f'pos_weight = {pos_weight.item():.3f} (neg={n_neg}, pos={n_pos})')

experiments = [
    ('BCE + aug',         nn.BCEWithLogitsLoss(),                                 loaders),
    ('Focal(g=2) + aug',  FocalLoss(gamma=2.0, alpha=0.5),                        loaders),
    ('BCE sin aug',       nn.BCEWithLogitsLoss(),                                 loaders_noaug),
    ('BCE + pos_weight',  nn.BCEWithLogitsLoss(pos_weight=pos_weight),            loaders),
]

ablation_rows = []
for tag, crit, ldrs in experiments:
    print(f'\n=== Ablation: {tag} ===')
    net = make_resnet_ft_clone().to(DEVICE)
    params = [p for p in net.parameters() if p.requires_grad]
    net, _ = train_model(net, ldrs, crit, epochs=EPOCHS_ABLATION, lr=1e-4,
                         params_to_optimize=params, patience=max(2, PATIENCE - 1), label=tag)
    val_m, *_ = evaluate(net, ldrs['val'])
    tst_m, *_ = evaluate(net, ldrs['test'])
    ablation_rows.append({'variant': tag,
                          'val_f1': val_m['f1'],  'val_auc': val_m['auc'],
                          'test_f1': tst_m['f1'], 'test_auc': tst_m['auc']})

df_abl = pd.DataFrame(ablation_rows)
print('\nResultados de ablación:'); print(df_abl.to_string(index=False))
```

## 10. Interpretabilidad — Grad-CAM (implementación propia)

```python
import cv2

class GradCAM:
    """Grad-CAM minimalista (Selvaraju et al., 2017) para CNN con salida escalar."""
    def __init__(self, model, target_layer):
        self.model = model.eval(); self.activations = None; self.gradients = None
        target_layer.register_forward_hook(self._save_act)
        target_layer.register_full_backward_hook(self._save_grad)
    def _save_act(self, mod, inp, out):     self.activations = out.detach()
    def _save_grad(self, mod, gin, gout):   self.gradients   = gout[0].detach()
    def __call__(self, x):
        x = x.to(DEVICE); self.model.zero_grad()
        logit = self.model(x).squeeze()
        logit.backward(retain_graph=True)
        weights = self.gradients.mean(dim=(2, 3), keepdim=True)
        cam = F.relu((weights * self.activations).sum(dim=1, keepdim=True))
        cam = F.interpolate(cam, size=(IMG_SIZE, IMG_SIZE), mode='bilinear', align_corners=False)
        cam = cam[0, 0].cpu().numpy()
        cam = (cam - cam.min()) / (cam.max() - cam.min() + 1e-8)
        return cam, torch.sigmoid(logit).item()

def denorm(img_tensor):
    mean = np.array(IMAGENET_MEAN).reshape(3, 1, 1)
    std  = np.array(IMAGENET_STD).reshape(3, 1, 1)
    return np.clip(img_tensor.cpu().numpy() * std + mean, 0, 1).transpose(1, 2, 0)

def overlay(img_rgb, cam, alpha=0.45):
    heat = cv2.applyColorMap(np.uint8(255 * cam), cv2.COLORMAP_JET)[:, :, ::-1] / 255.0
    return np.clip(alpha * heat + (1 - alpha) * img_rgb, 0, 1)

cam_engine = GradCAM(resnet, resnet.layer4[-1])

fig, axes = plt.subplots(2, 4, figsize=(14, 7))
samples = pd.concat([
    df_test[df_test['label'] == 0].sample(4, random_state=SEED),
    df_test[df_test['label'] == 1].sample(4, random_state=SEED),
]).reset_index(drop=True)
for i, (_, row) in enumerate(samples.iterrows()):
    img = Image.open(row['path']).convert('RGB')
    x = eval_tf(img).unsqueeze(0)
    cam, prob = cam_engine(x)
    ax = axes[i // 4, i % 4]
    ax.imshow(overlay(denorm(x[0]), cam))
    ax.set_title(f"real={'cracked' if row['label']==1 else 'normal'} | p={prob:.2f}")
    ax.axis('off')
plt.suptitle('Grad-CAM sobre ResNet-50', fontsize=14); plt.tight_layout(); plt.show()
```

## 11. Análisis cualitativo de fallos (al menos 5 ejemplos mal clasificados)

```python
import shutil
from pathlib import Path

ERRORS_DIR = Path("outputs/errors")
FP_DIR = ERRORS_DIR / "false_positives"
FN_DIR = ERRORS_DIR / "false_negatives"
for d in [FP_DIR, FN_DIR]:
    d.mkdir(parents=True, exist_ok=True)

y_pred_rn = (y_prob_rn >= 0.5).astype(int)
all_errors = pd.DataFrame({
    "path": paths_rn,
    "true": y_true_rn.astype(int),
    "pred": y_pred_rn,
    "prob": y_prob_rn,
})
all_errors = all_errors[all_errors["true"] != all_errors["pred"]].copy()
all_errors["tipo_error"]       = np.where(all_errors["pred"] == 1, "FP", "FN")
all_errors["confidence_error"] = np.where(
    all_errors["pred"] == 1, all_errors["prob"], 1 - all_errors["prob"]
)
all_errors = all_errors.sort_values("confidence_error", ascending=False)

log_path = ERRORS_DIR / "error_log.csv"
all_errors[["path", "true", "pred", "prob", "tipo_error"]].rename(columns={
    "path": "ruta_imagen", "true": "etiqueta_real",
    "pred": "etiqueta_predicha", "prob": "probabilidad_predicha",
}).to_csv(log_path, index=False)
fp_count = (all_errors["tipo_error"] == "FP").sum()
fn_count = (all_errors["tipo_error"] == "FN").sum()
print(f"error_log.csv guardado en {log_path}  ({len(all_errors)} errores totales)")
print(f"  FP: {fp_count}  |  FN: {fn_count}")

N_SAVE = 5
for tipo, subset_dir in [("FP", FP_DIR), ("FN", FN_DIR)]:
    subset = all_errors[all_errors["tipo_error"] == tipo].head(N_SAVE)
    for _, row in subset.iterrows():
        src = Path(row["path"])
        fname = f"{tipo.lower()}_prob{row['prob']:.3f}_{src.stem}{src.suffix}"
        shutil.copy(src, subset_dir / fname)
    print(f"{tipo}: {len(subset)} imagen(es) guardada(s) en {subset_dir}")

top5 = all_errors.head(5)
fig, axes = plt.subplots(1, 5, figsize=(18, 4))
for ax, (_, row) in zip(axes, top5.iterrows()):
    img = Image.open(row["path"]).convert("RGB")
    x = eval_tf(img).unsqueeze(0)
    cam, prob = cam_engine(x)
    ax.imshow(overlay(denorm(x[0]), cam))
    tag_true = "cracked" if row["true"] == 1 else "normal"
    tag_pred = "cracked" if row["pred"] == 1 else "normal"
    ax.set_title(f"real={tag_true}\npred={tag_pred} ({prob:.2f})\n[{row['tipo_error']}]")
    ax.axis("off")
plt.suptitle("Top 5 errores con mayor confianza -- ResNet-50", fontsize=14)
plt.tight_layout()
plt.savefig("figures/fig7_error_analysis.png", dpi=150, bbox_inches="tight")
plt.show()
print("fig7_error_analysis.png guardado en figures/")
```

## 12. Resumen final

- `TireCNN` (entrenada desde cero) alcanza una línea base razonable pero limitada por el tamaño del corpus.
- `ResNet-50` con fine-tuning gradual obtiene la mejor F1 y AUC en test.
- En la ablación, el aumento de datos es el factor que más aporta; las técnicas de desbalance (focal, pos_weight) no producen ganancia medible porque el desbalance es leve.
- Grad-CAM confirma que las activaciones se concentran sobre grietas/oxidación; los errores más confiados se explican por letras moldeadas en el flanco, sombras duras y, en un caso, etiquetado dudoso en el dataset original.

```python
# --- Volcado de métricas + historial + ablación a un único JSON ---
def _cm(y_true, y_prob, thr=0.5):
    cm = confusion_matrix(y_true, (y_prob >= thr).astype(int))
    return {'tn': int(cm[0, 0]), 'fp': int(cm[0, 1]),
            'fn': int(cm[1, 0]), 'tp': int(cm[1, 1])}

# Búsqueda de umbral óptimo (maximiza F1 en VALIDACIÓN, evita usar test)
_, y_val_rn, p_val_rn, _ = evaluate(resnet, loaders['val'])
_, y_val_cnn, p_val_cnn, _ = evaluate(model_cnn, loaders['val'])

def best_threshold(y_true, y_prob):
    grid = np.linspace(0.05, 0.95, 91)
    f1s = [f1_score(y_true, (y_prob >= t).astype(int), zero_division=0) for t in grid]
    i = int(np.argmax(f1s))
    return float(grid[i]), float(f1s[i])

thr_cnn, f1_val_cnn = best_threshold(y_val_cnn, p_val_cnn)
thr_rn,  f1_val_rn  = best_threshold(y_val_rn,  p_val_rn)
print(f'Umbral óptimo (val) — TireCNN  : {thr_cnn:.2f} (F1 val={f1_val_cnn:.3f})')
print(f'Umbral óptimo (val) — ResNet-50: {thr_rn:.2f}  (F1 val={f1_val_rn:.3f})')

metrics_cnn_05  = compute_metrics(y_true_cnn, y_prob_cnn, threshold=0.5)
metrics_cnn_opt = compute_metrics(y_true_cnn, y_prob_cnn, threshold=thr_cnn)
metrics_rn_05   = compute_metrics(y_true_rn,  y_prob_rn,  threshold=0.5)
metrics_rn_opt  = compute_metrics(y_true_rn,  y_prob_rn,  threshold=thr_rn)

bundle = {
    'environment': {
        'device': str(DEVICE), 'on_gpu': ON_GPU,
        'in_kaggle': IN_KAGGLE, 'in_colab': IN_COLAB,
        'torch': torch.__version__, 'seed': SEED, 'img_size': IMG_SIZE, 'batch': BATCH,
        'epochs': {'cnn': EPOCHS_CNN, 'resnet_head': EPOCHS_RESNET_HEAD,
                   'resnet_ft': EPOCHS_RESNET_FT, 'ablation': EPOCHS_ABLATION,
                   'patience': PATIENCE},
    },
    'dataset': {
        'train': int(len(df_tr)), 'val': int(len(df_val)), 'test': int(len(df_test)),
        'class_map': MAP,
    },
    'tirecnn': {
        'n_params': int(sum(p.numel() for p in model_cnn.parameters())),
        'history': hist_cnn.to_dict(orient='records'),
        'test_thr0.5':  {**metrics_cnn_05,  'cm': _cm(y_true_cnn, y_prob_cnn, 0.5)},
        'test_thr_opt': {'threshold': thr_cnn,
                        **metrics_cnn_opt, 'cm': _cm(y_true_cnn, y_prob_cnn, thr_cnn)},
    },
    'resnet50': {
        'n_params_trainable': int(sum(p.numel() for p in resnet.parameters() if p.requires_grad)),
        'history_head': hist_r1.to_dict(orient='records'),
        'history_ft':   hist_r2.to_dict(orient='records'),
        'test_thr0.5':  {**metrics_rn_05,  'cm': _cm(y_true_rn, y_prob_rn, 0.5)},
        'test_thr_opt': {'threshold': thr_rn,
                        **metrics_rn_opt, 'cm': _cm(y_true_rn, y_prob_rn, thr_rn)},
    },
    'ablation': df_abl.to_dict(orient='records'),
}

out_path = (WORK_DIR / 'metrics.json').resolve()
out_path.write_text(json.dumps(bundle, indent=2, default=float))
print(f'\nMétricas exportadas a: {out_path}')
print(f'Tamaño: {out_path.stat().st_size} bytes')
print('\n--- Resumen final (test) ---')
print(f"TireCNN   thr=0.5  F1={metrics_cnn_05['f1']:.3f}  AUC={metrics_cnn_05['auc']:.3f}")
print(f"TireCNN   thr={thr_cnn:.2f} F1={metrics_cnn_opt['f1']:.3f}")
print(f"ResNet-50 thr=0.5  F1={metrics_rn_05['f1']:.3f}  AUC={metrics_rn_05['auc']:.3f}")
print(f"ResNet-50 thr={thr_rn:.2f} F1={metrics_rn_opt['f1']:.3f}")
```
